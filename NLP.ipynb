```python
from bs4 import BeautifulSoup
import nltk
nltk.download("all")
import pandas as pd
from nltk.tokenize import word_tokenize, RegexpTokenizer, sent_tokenize
import urllib.request
import numpy as np

input_data=pd.read_excel("Input.xlsx", index_col=0)
output_data=pd.read_excel("Output Data Structure (1).xlsx",index_col=0)
output_data.head()
input_data.head()
pages=[]
pages=output_data['URL'].values.tolist()
positive_list=[]
negative_list=[]
polarity_list=[]
subjectivity_list=[]
sent_length_list=[]
perc_comp_list=[]
fog_index_list=[]
word_per_list=[]
comp_count_list=[]
word_count_list=[]
syll_word_list=[]
pron_list=[]
word_length_list=[]
for x in range(len(pages)):
  #extracting data
  try:
    page = urllib.request.urlopen(pages[x])
  except urllib.error.HTTPError as e:
    continue
  else:
    print("successful")
  soup = BeautifulSoup(page, 'html.parser')
  string=[]
  for para in soup.find_all("p"):
      aka=para.get_text()
      tokenizer= RegexpTokenizer(r'\w+')
      temp=tokenizer.tokenize(aka)
      string=(string+temp)
  print(string)

  string_string = ' '.join([str(elem) for i,elem in enumerate(string)])



  #cleaning the list
  my_file = open("/content/StopWords_Auditor.txt", encoding="utf-8")
  data1 = my_file.read()


  datas1 = data1.split("\n")

  my_file.close()

  my_file = open("/content/StopWords_Currencies.txt", encoding="utf-8")
  data2 = my_file.read()


  datas2 = data2.split("\n")

  my_file.close()

  my_file = open("/content/StopWords_DatesandNumbers.txt", encoding="utf-8")
  data3 = my_file.read()


  datas3 = data3.split("\n")

  my_file.close()

  my_file = open("/content/StopWords_Generic.txt", encoding="utf-8")
  data4 = my_file.read()


  datas4 = data4.split("\n")

  my_file.close()

  my_file = open("/content/StopWords_GenericLong.txt", encoding="utf-8")
  data5 = my_file.read()


  datas5 = data5.split("\n")

  my_file.close()

  my_file = open("/content/StopWords_Geographic.txt", encoding="utf-8")
  data6 = my_file.read()


  datas6 = data6.split("\n")

  my_file.close()

  my_file = open("/content/StopWords_Names.txt", encoding="utf-8")
  data7 = my_file.read()


  datas7 = data7.split("\n")

  my_file.close()

  cleaning_data=datas1+datas2+datas3+datas4+datas5+datas6+datas7
  len(cleaning_data)


  cleaned_string=[i for i in string if i not in cleaning_data]
  words=len(string)
  print(len(cleaned_string))
  print(cleaned_string)

  #negative word count
  my_file = open("/content/negative-words.txt", encoding="utf-8" )
  neg_data = my_file.read()
  negatives = neg_data.split("\n")
  my_file.close()
  negatives_string=[i for i in cleaned_string if i in negatives]
  neg_score=len(negatives_string)
  negative_list.append(neg_score)

  #postive word count
  my_file = open("/content/positive-words.txt", encoding="utf-8")
  pos_data = my_file.read()
  positives = pos_data.split("\n")
  my_file.close()
  positives_string=[i for i in cleaned_string if i in positives]
  pos_score=len(positives_string)
  positive_list.append(pos_score)


  #polarity score & subjectivitiy
  Polarity_Score = (pos_score - neg_score) / ((pos_score + neg_score) + 0.000001)
  print(Polarity_Score)
  polarity_list.append(Polarity_Score)
  Subjectivity_Score = (pos_score + neg_score)/ ((len(cleaned_string)) + 0.000001)
  print(Subjectivity_Score)
  subjectivity_list.append(Subjectivity_Score)



  #avg sent length
  string2=[]
  for para in soup.find_all("p"):
      aka=para.get_text()
      temp=sent_tokenize(aka)
      string2=(string2+temp)
  print(string2)
  word_per_sent=len(string)/len(string2)
  print(word_per_sent)
  sent_length_list.append(word_per_sent)
  word_per_list.append(word_per_sent)


  #complex word %age
  count = 0

  for myword in string:
      d = {}.fromkeys('aeiou',0)
      haslotsvowels = False
      for x in myword.lower():
          if x in d:
              d[x] += 1
      for q in d.values():
          if q > 2:
              haslotsvowels = True
      if haslotsvowels:
          count += 1
  complex_percent=(count/words)
  print(complex_percent)
  perc_comp_list.append(complex_percent)
  #fog_index
  fog_index=(word_per_sent+complex_percent)*0.4
  print(fog_index)
  fog_index_list.append(fog_index)
  #complex word count
  print(count)
  #word count
  word_count=len(cleaned_string)
  word_count_list.append(word_count)

  #syllables per word
  import re

  def count_syllables(word):
      return len(
          re.findall('(?!e$)[aeiouy]+', word, re.I) +
          re.findall('^[^aeiouy]*e$', word, re.I)
      )
  print(count_syllables(string_string))

  syll_per_word=count_syllables(string_string)/len(string)
  print(syll_per_word)
  syll_word_list.append(syll_per_word)






  #personal pronouns(our added from the initial list of personal pronouns)
  I=string_string.count(" I")
  print(type(I))
  our=string_string.count(" our")
  Our=string_string.count(" Our")
  we=string_string.count(" we")
  We=string_string.count(" We")
  my=string_string.count(" my")
  My=string_string.count(" My")
  ours=string_string.count(" ours")
  Ours=string_string.count(" Ours")
  us=string_string.count(" us")
  Us=string_string.count(" Us")
  personal=I+we+We+my+My+ours+Ours+us+Us+our+Our
  pron_list.append(personal)



  #print no. of characters/word
  print(string_string)
  lst=[x for x in string_string]
  res = []
  for ele in lst:
      if ele.strip():
          res.append(ele)
  char_per_word=len(res)/len(string)
  print(char_per_word)
  word_length_list.append(char_per_word)

```
